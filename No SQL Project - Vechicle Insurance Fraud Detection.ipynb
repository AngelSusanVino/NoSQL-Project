### Encoding Categorical Values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Identify the columns with mixed data types (e.g., both strings and integers)
mixed_type_columns = df.select_dtypes(include=['object']).columns

# Convert the mixed-type columns to strings
for col in mixed_type_columns:
    df[col] = df[col].astype(str)

# Perform Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in mixed_type_columns:
    df[col] = le.fit_transform(df[col])

# Now you can use the encoded DataFrame
print(df)
```

### Feature Selection Using RFE

```python
Y = df['FraudFound_P']
X = df.drop(['FraudFound_P'], axis=1)
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import AdaBoostClassifier
```

```python
model=AdaBoostClassifier(random_state=0, n_estimators=100)
rfe = RFE(model,n_features_to_select=20,step=1)
fit = rfe.fit(X, Y)
print("Num Features: %s" % (fit.n_features_))
print("Selected Features: %s" % (fit.support_))
print("Feature Ranking: %s" % (fit.ranking_))
```

```python
from numpy import array
features = array(X.columns)
print("All features:")
print(features)
filter = rfe.support_

print("Selected features:")
print(features[filter]) 
```

## Under Sampling Using Nearmiss

```python
import imblearn
from imblearn.under_sampling import NearMiss
nm = NearMiss()
x_nm, y_nm = nm.fit_resample(x, y)
```

```python
x_nm.shape
```

```python
y_nm.shape
```

```python
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_nm, y_nm, test_size=0.20, random_state=12)
```

```python
DT_mod_norm.fit(x_train1, y_train1) 
DT_mod_norm.score(x_test1,y_test1)
```

```python
print(classification_report(y_test1, DT_y_predicted1))
scores3 = cross_val_score(DT_mod_norm, x_nm, y_nm, cv=10, scoring='f1')
DT_fsc = scores3.mean()
print('f1: %0.2f' % DT_fsc)
```

### KNN Under Sampling
